## Basics

```python
from bark import SAMPLE_RATE, generate_audio, preload_models
from IPython.display import Audio
preload_models()
```

```python
# 0 : 기본 여성 목소리
# 6 : 중후한 교수님
# 7 : 걸걸한 교수님
# 8 : 나이 많으신 교수님
# 9 : 스프치 강사
number = 7 # 0 ~ 9
text_prompt = """
    여러분의 제품에 API를 연동해 보고 싶으신가요?
"""
# text_prompt = "250쿼리를 제공 합니다." # Good : 1, 4
# https://github.com/suno-ai/bark/issues/380#issuecomment-1625127198
audio_array = generate_audio(text_prompt, history_prompt=f"v2/ko_speaker_{number}")
Audio(audio_array, rate=SAMPLE_RATE)
```
